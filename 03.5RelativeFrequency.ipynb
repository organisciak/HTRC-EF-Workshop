Now we'll look at a look using relative frequencies. Relative frequencies are one way at looking at top words, through their proportional counts. Books have different lengths, so the nominal count of any given word will vary between books, so relative frequencies give us a way to compare two or more books.

```python
%matplotlib inline
from htrc_features import FeatureReader
```

## Kreacher Example

```python
vol = FeatureReader(ids=['uc1.32106019703807']).first()
tl = vol.tokenlist(pos=True, pages=True, case=True)
kreacher = tl.loc[(slice(None), slice(None), 'Kreacher','NNP'),].reset_index()
kreacher.plot(x='page', y='count')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
subset = tl.loc[(slice(None), slice(None), ('Harry', 'Ron', 'Hermione', 'Dumbledore', 'Kreacher')),].reset_index()
g = sns.FacetGrid(subset, col="token", col_wrap=3)
g = g.map(plt.scatter, "page", "count")
```

## Relative Frequencies

```python
fr = FeatureReader('data/sample-file1.json.bz2')
vol = fr.first()
tokens = vol.tokenlist(pages=False, pos=False, case=False)
tokens = tokens.loc['body'] # Only focus on section='body'
```

```python
tokens['rel_freq'] = tokens['count'] / tokens['count'].sum()
tokens = tokens.sort_values('rel_freq', ascending=False)
tokens.head(15).plot(y='rel_freq', kind='bar', title='Common token frequencies in ' + vol.title)
```

```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
subset = tokens[~tokens.index.isin(stopwords.words('english')) & tokens.index.str.isalpha()]
subset.head(15).plot(y='rel_freq', kind='bar', title='Common token frequencies in ' + vol.title)
```
